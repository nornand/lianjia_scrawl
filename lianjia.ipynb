```python
def get_page_num(url):
    r = requests.get(url,headers=headers)
    bs = BeautifulSoup(r.text,'lxml')
    page_num = math.ceil(int(bs.find(attrs={'class':'total fl'}).span.string.strip())/30)
    return page_num
```

```python
def get_house(url):
    pd_house_deal = pd.DataFrame()
    pd_house_detail = pd.DataFrame()
    r = requests.get(url, headers=headers)
    #print(r.status_code)
    #print(r.text)
    soup = BeautifulSoup(r.text, 'lxml')
    
    dict_cur_house ={}
    #list_content = soup.find_all(attrs={'class': 'listContent'})
    for content in soup.find(attrs={'class': 'listContent'}).contents:
        dict_cur_house['house_name'] = content.find(attrs={'class':'title'}).string.split(' ')[0]
        dict_cur_house['remark'] = ''
        try:  
            #print(dict_cur_house['house_name'])
            dict_cur_house['house_type'] = content.find(attrs={'class':'title'}).string.split(' ')[1]
            dict_cur_house['house_area'] = content.find(attrs={'class':'title'}).string.split(' ')[2]
        

            dict_cur_house['house_ori'] = content.find(attrs={'class':'houseInfo'}).contents[1].split(' | ')[0]
            dict_cur_house['house_deco'] = content.find(attrs={'class':'houseInfo'}).contents[1].split(' | ')[1]


            dict_cur_house['deal_date'] = content.find(attrs={'class':'dealDate'}).string

            dict_cur_house['totle_price'] = content.find(attrs={'class':'totalPrice'}).contents[0].string

            dict_cur_house['totle_price_unit'] = content.find(attrs={'class':'totalPrice'}).contents[1].string

            dict_cur_house['aver_price'] = content.find(attrs={'class':'unitPrice'}).contents[0].string
            dict_cur_house['aver_price_unit'] =  content.find(attrs={'class':'unitPrice'}).contents[1].string

            dict_cur_house['house_floor'] = content.find(attrs={'class':'positionInfo'}).contents[1].split(' ')[0]
            dict_cur_house['house_create'] = content.find(attrs={'class':'positionInfo'}).contents[1].split(' ')[1]

            dict_cur_house['house_info'] = ''
            if content.find(attrs={'class':'dealHouseTxt'}):
                dict_cur_house['house_info'] = '|'.join([i.string for i in content.find(attrs={'class':'dealHouseTxt'}).contents])

            dict_cur_house['list_price'] = content.find(attrs={'class':'dealCycleTxt'}).contents[0].string
            dict_cur_house['deal_period'] = content.find(attrs={'class':'dealCycleTxt'}).contents[1].string
            if content.find(attrs={'class':'agent_chat_btn'}):
                dict_cur_house['house_code'] = content.find(attrs={'class':'agent_chat_btn'}).attrs['data-lj_action_house_code']
                dict_cur_house['agent_id'] = content.find(attrs={'class':'agent_chat_btn'}).attrs['data-lj_action_agent_id']
                dict_cur_house['agent_name'] = content.find(attrs={'class':'agent_chat_btn'}).attrs['data-lj_action_agent_name']
            pd_house_deal = pd_house_deal.append(dict_cur_house,ignore_index=True)
        
        except IndexError as e:
            dict_cur_house['remark'] = e
        
        dict_house_detail = get_house_detail(content.find(attrs={'class':'title'}).a['href'])
        pd_house_detail = pd_house_detail.append(dict_house_detail,ignore_index=True)
        
    return pd_house_deal,pd_house_detail
```

```python
def get_house_detail(url):
    #print(url)
    dict_house_detail = {}
    r = requests.get(url,headers=headers)
    bs = BeautifulSoup(r.text,'lxml')
    base_content = bs.find(attrs={'class':'base'}).ul.contents
    dict_house_detail['url'] = url
    dict_house_detail['remark'] = ''
    try:
        dict_house_detail['house_style'] = base_content[0].contents[1].string.strip()
        dict_house_detail['house_floor'] = base_content[1].contents[1].string.strip()
        dict_house_detail['house_area'] = base_content[2].contents[1].string.strip()
        dict_house_detail['house_type'] = base_content[3].contents[1].string.strip()
        dict_house_detail['area_in'] = base_content[4].contents[1].string.strip()
        dict_house_detail['build_type'] = base_content[5].contents[1].string.strip()
        dict_house_detail['house_ori'] = base_content[6].contents[1].string.strip()
        dict_house_detail['build_year'] = base_content[7].contents[1].string.strip()
        dict_house_detail['house_dec'] = base_content[8].contents[1].string.strip()
        dict_house_detail['build_struct'] = base_content[9].contents[1].string.strip()
        dict_house_detail['warm_type'] = base_content[10].contents[1].string.strip()
        dict_house_detail['elevator_prop'] = base_content[11].contents[1].string.strip()
        dict_house_detail['property_year'] = base_content[12].contents[1].string.strip()
        dict_house_detail['elevator'] = base_content[13].contents[1].string.strip()

        tran_content = bs.find(attrs={'class':'transaction'}).ul.contents
        dict_house_detail['house_code'] = tran_content[0].contents[1].string.strip()
        dict_house_detail['tran_type'] = tran_content[1].contents[1].string.strip()
        dict_house_detail['hang_date'] = tran_content[2].contents[1].string.strip()
        dict_house_detail['house_use'] = tran_content[3].contents[1].string.strip()
        dict_house_detail['bought_year'] = tran_content[4].contents[1].string.strip()
        dict_house_detail['own_type'] = tran_content[5].contents[1].string.strip()

        dict_house_detail['trans_amount'] = bs.find(attrs={'class':'record_price'}).string
        dict_house_detail['trans_unit'] = bs.find(attrs={'class':'record_detail'}).string.split(',')[0]
        dict_house_detail['trans_date'] = bs.find(attrs={'class':'record_detail'}).string.split(',')[1]

    except IndexError as e:
        dict_house_detail['remark'] = e
    
    
    return dict_house_detail
```
